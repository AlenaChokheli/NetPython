```python
#Задание 2. Работа с выбросами

"""В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. 
Принять и обосновать решение о дальнейшей работе с ними."""
```

```python
q1 = df.Age.quantile(0.25)
q3 = df.Age.quantile(0.75)
iqr = q3 - q1
lower_Age = q1 - (1.5 * iqr) 
upper_Age = q3 + (1.5 * iqr)
remove_Age = df[df.Age.between(lower_Age, upper_Age, inclusive=True)].sort_values('Age')
# что это за выброс?
pd.concat([df, remove_Age]).drop_duplicates(keep=False)
```

```python
q1 = df.rectal_temperature.quantile(0.25)
q3 = df.rectal_temperature.quantile(0.75)
iqr = q3 - q1
lower_rectal_temperature = q1 - (1.5 * iqr) 
upper_rectal_temperature = q3 + (1.5 * iqr)
remove_rectal_temperature = df[df.rectal_temperature.between(lower_rectal_temperature, upper_rectal_temperature, inclusive=True)].sort_values('rectal_temperature')
# что это за выброс?
pd.concat([df, remove_rectal_temperature]).drop_duplicates(keep=False)
```

```python
q1 = df.pulse.quantile(0.25)
q3 = df.pulse.quantile(0.75)
iqr = q3 - q1
lower_pulse = q1 - (1.5 * iqr) 
upper_pulse = q3 + (1.5 * iqr)
remove_pulse = df[df.pulse.between(lower_pulse, upper_pulse, inclusive=True)].sort_values('pulse')
# что это за выброс?
pd.concat([df, remove_pulse]).drop_duplicates(keep=False)
```

```python
q1 = df.capillary_refill_time.quantile(0.25)
q3 = df.capillary_refill_time.quantile(0.75)
iqr = q3 - q1
lower_capillary_refill_time = q1 - (1.5 * iqr) 
upper_capillary_refill_time = q3 + (1.5 * iqr)
remove_capillary_refill_time = df[df.capillary_refill_time.between(lower_capillary_refill_time, upper_capillary_refill_time, inclusive=True)].sort_values('capillary_refill_time')
# что это за выброс?
pd.concat([df, remove_capillary_refill_time]).drop_duplicates(keep=False)
```

```python
df = pd.read_csv('horse_data.csv',na_values='?')
df.columns = ['surgery', 'Age', 'Hospital_Number', 'rectal_temperature', 'pulse', 'respiratory_rate', 
              'temperature_of_extremities', 'peripheral_pulse', 'mucous_membranes','capillary_refill_time', 
              'pain', 'peristalsis', 'abdominal_distension', 'nasogastric_tube', 'nasogastric_reflux', 
              'nasogastric_reflux_PH', 'rectal_examination_feces', 'abdomen', 'packed_cell_volume', 
              'total_protein', 'abdominocentesis_appearance', 'abdomcentesis_total_protein', 'outcome', 
              'surgical_lesion', 'type_of_lesion1', 'type_of_lesion2', 'type_of_lesion3', 'cp_data']
horse_data_ = df.iloc[:, [0,1,2,3,4,8,9,12]]

```

```python
horse_data_.dropna(thresh=6).info() #удалила строки в которых нет хотя бы 6 записей
```

```python
horse_data_.rectal_temperature.fillna(df.rectal_temperature.median(),inplace=True)
horse_data_.rectal_temperature.fillna(df.rectal_temperature.median()).isna().sum()
```

```python
horse_data_.pulse.fillna(df.pulse.median(),inplace=True)
horse_data_.pulse.fillna(df.pulse.median()).isna().sum()
```

```python
horse_data_.mucous_membranes.fillna(df.mucous_membranes.mode()[0],inplace=True)
horse_data_.mucous_membranes.fillna(df.mucous_membranes.mode()[0]).isna().sum()
```

```python
horse_data_.capillary_refill_time.fillna(df.capillary_refill_time.mode()[0],inplace=True)
horse_data_.capillary_refill_time.fillna(df.capillary_refill_time.mode()[0]).isna().sum()
```

```python
# мы не можем исключать, что операция была показана тем лошадям, у которых было сильнее вздутие живота

horse_data_.abdominal_distension.fillna(horse_data_.groupby('surgery').abdominal_distension.transform('median'), inplace=True)
print(horse_data_.groupby('surgery').abdominal_distension.median())
```

```python
#процент пустых значений
for col in horse_data_.columns:
    pct_missing = horse_data_[col].isna().mean()
    print(f'{col} - {pct_missing :.1%}')
```
