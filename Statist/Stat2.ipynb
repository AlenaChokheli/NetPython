```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
df = pd.read_csv('2019.csv')
df = df. rename(columns={'Country or region':'Country_or_region'})
df = df. rename(columns={'GDP per capita':'GDP_per_capita'})
df = df. rename(columns={'Social support':'Social_support'})
df = df. rename(columns={'Healthy life expectancy':'Healthy_life_expectancy'})
df = df. rename(columns={'Freedom to make life choices':'Freedom_to_make_life_choices'})
df = df. rename(columns={'Perceptions of corruption':'Perceptions_of_corruption'})
df = df. rename(columns={'Overall rank':'Overall_rank'})
df.head(20)
```

```python
df.info()
```

```python
df.groupby('Country_or_region')['Score'].max().sort_values(ascending=False).head(20).plot(kind='barh', color='green')
plt.title('Top 20 happiest countries (2019)')
```

```python
print('Данная диаграмма наглядно показывает итоги опроса населения и топ 20 самых счастливых стран')
```

```python
ax = sns.regplot(data=df, x='Freedom_to_make_life_choices', y='Score', color='green', marker='+')
ax.set_title('Correlation between Freedom_to_make_life_choices and Score')
```

```python
print('Мы можем предположить, что уровень жизни напрямую зависит от свободы выбора в стране')
```

```python
df.plot(kind='scatter', 
        x='Social_support', 
        y='Generosity', 
        title='Взаимосвязь социальной поддержки населения и благотворительность',
        color='green'
       )
```

```python
print('По данной диаграмее можно сделать вывод, что чем больше социальная поддержка населения, тем чаще люди участвуют в благотворительности.')
```

```python
sns.pairplot(df[['GDP_per_capita', 'Social_support', 'Healthy_life_expectancy']])
```

```python
print('Мы можем посмотреть на одной картинке, как связаны между собой ВВП на душу населения, социальная поддержка и ожидаемая продолжительность жизни. Можно выдвинуть гипотезу, что чем выше ВВП на душу населения, увеличитвается соц. поддержка, в связи с чем увеличивается продолжительность жизни.')
```
